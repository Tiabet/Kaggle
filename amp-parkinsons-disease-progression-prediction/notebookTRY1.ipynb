```python
import pandas as pd
import numpy as np

train_clinical_data = pd.read_csv('train_clinical_data.csv')
train_peptides = pd.read_csv('train_peptides.csv')
train_proteins = pd.read_csv('train_proteins.csv')
```

```python
train_clinical_data['updrs_4'].fillna(0, inplace=True)
train_clinical_data['upd23b_clinical_state_on_medication'].fillna('Off', inplace=True)

train_clinical_data_pivot = pd.pivot_table(train_clinical_data, values=['updrs_1', 'updrs_2', 'updrs_3', 'updrs_4', 'upd23b_clinical_state_on_medication'], index=['patient_id', 'visit_month'])
```

```python
train_peptides['PeptideAbundance'] = pd.to_numeric(train_peptides['PeptideAbundance'], errors='coerce')
train_peptides = train_peptides.groupby(['visit_id', 'patient_id', 'UniProt'])['PeptideAbundance'].mean().reset_index()
train_peptides_pivot = pd.pivot_table(train_peptides, values='PeptideAbundance', index=['patient_id'], columns='UniProt')

train_proteins = train_proteins[['patient_id', 'visit_month', 'UniProt', 'NPX']]
train_proteins = train_proteins.drop_duplicates(subset=['patient_id', 'visit_month', 'UniProt'])
```

```python
merged_df = pd.merge(train_clinical_data, train_peptides, on='visit_id', how='inner')
```

```python
merged_df = pd.merge(train_clinical_data, train_proteins, on='patient_id', how='inner')
```
